## Case 1: without solar and gas costs, only CO2 store cost

```python
import pypsa
import numpy as np

# Create network
n = pypsa.Network()
n.set_snapshots(range(7))  # Adjust as needed

# --- Buses ---
n.add(
    "Bus",
    "elec",
    carrier="AC",
)

n.add(
    "Generator",
    "solar_gen",
    bus="elec",
    carrier="solar",
    p_nom_extendable=True,
    capital_cost=0,
    p_max_pu=[1,1,1,1,0,0,0]
)


n.add(
    "Bus",
    "co2",
    carrier="CO2",
)


n.add(
    "Bus",
    "gas",
    carrier="gas",
)

n.add(
    "Generator",
    "gas_gen",
    bus="gas",
    carrier="gas",
    p_nom_extendable=True,
    capital_cost=0
)

n.add(
    "Load",
    "load",
    bus="elec",
    p_set=1,
)

n.add(
        "Link",
        "gas",
        bus0='gas',
        bus1='elec',
        bus2='co2',
        carrier="gas",
        p_nom_extendable=True,
        capital_cost=0,  # EUR/MW_H2/a
        lifetime=20,
        efficiency=1,
        efficiency2=1,
)


#----stores----#

n.add(
        "Store",
        "co2_store",
        bus='co2',
        carrier="co2",
        e_nom_extendable=True,
        capital_cost=100,  # EUR/MW_H2/a
        lifetime=20,
        e_nom_max=8, ## or 8
    )

# --- Solve ---
n.optimize(solver_name="gurobi")  # Uncomment to solve
n.statistics()  # View results
```

```python
n.buses_t.marginal_price
```

```python
n.generators_t.p
```
